##### Embeddings
Textos convertidos em vetores (listas de números) 

Lembrando de instalar o modelo:

```bash
ollama pull nomic-embed-text
```

````python
from langchain_ollama import OllamaEmbeddings
````

````python
embedding_model = OllamaEmbeddings(
    model="nomic-embed-text",
)
````

````python
embeddings = embedding_model.embed_documents([
    "texto a", "texto b", "texto c", "um texto longo"
])
````

##### Cada posicao tem tamanhos fixos (embending[0 em diante])

````python
len(embeddings), len(embeddings[0]), len(embeddings[1]), len(embeddings[2])
````

````python
min(embeddings[0]), max(embeddings[0]), min(embeddings[1]), max(embeddings[1]), min(embeddings[2]), max(embeddings[2])
````

##### Podemos fazer a comparação vetorial. Para isso podemos fazer a multiplicação. 
Quanto mais alto for, mais proximos são semanticamentes.

````python
import numpy as np
````

````python
np.dot(embeddings[0], embeddings[1])
````

````python
np.dot(embeddings[1], embeddings[2])
````

````python
np.dot(embeddings[2], embeddings[3])
````

````python
for i in range(len(embeddings)):
    for j in range(len(embeddings)):
        print(f"{round(np.dot(embeddings[i], embeddings[j]), 2):.2f}-{(i, j)}", end="|")
    print()
````
