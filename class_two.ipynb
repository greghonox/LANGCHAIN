### Aqui começamos a estudar as CHAIN

```python
from langchain_community.llms import Ollama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
from dotenv import load_dotenv

load_dotenv()
```

```python
chat_model = Ollama(model="minimax-m2.5:cloud", temperature=0)
```

```python
prompt_template = ChatPromptTemplate.from_template("Você é um assistente muito educado e amigável e vai responder as seguintes perguntas: {pergunta}")
```

```python
chain = prompt_template | chat_model

response = chain.invoke({"pergunta": "Como você está?"})

print(response)
```

```python
chain = prompt_template | chat_model | StrOutputParser()

response = chain.invoke({"pergunta": "você pode dominar o mundo?"})

print(response)
```

### Tabela de Componentes e a Sequencia 

| Component | Tipo de Entrada               | Tipo de Saída               |
|---------------|-----------------------------------|----------------------------------|
| Prompt        | Dicionário                        | PromptValue                    |
| ChatModel     | String única, lista de mensagens de chat ou PromptValue | Mensagem de Chat |
| LLM           | String única, lista de mensagens de chat ou PromptValue | String |
| OutputParser  | Saída de um LLM ou ChatModel       | Depende do parser                |
| Retriever     | String única                      | Lista de Documentos              |
| Tool          | String única ou dicionário, dependendo da ferramenta | Depende da ferramenta            |

### Ordem de Uso de um Chain
A ordem típica de uso é a seguinte:

1. Entrada (Input):
   - Você fornece uma entrada (ex.: uma pergunta ou texto).
   - Exemplo: "Qual a capital da França?".

2. Prompt:
   - O prompt é criado com base na entrada para guiar o modelo.
   - Exemplo: "Responda à pergunta: {question}".

3. Retiwer (Opcional):
   - Se o Chain precisar de informações externas, o Retiwer busca dados relevantes.
   - Exemplo: Buscar informações sobre a França em um banco de dados ou vetores.

4. LLM (Modelo de Linguagem):
   - O modelo processa o prompt (e os dados do Retiwer, se aplicável) e gera uma resposta.

5. Output Parser (Opcional):
   - Se o Chain precisar extrair informações específicas da saída do modelo, o Output Parser converte a resposta em um formato útil.

6. Saída (Output):
   - A resposta final é retornada ao usuário.

---

```python
prompt = ChatPromptTemplate.from_template("Criar um texto que seja traduzido para o {idioma}: {texto}")
output_parser = StrOutputParser()

chain = prompt | chat_model | output_parser

texto = "Jesus voltará em breve!!!"
response = chain.invoke({"texto": texto, "idioma":"japonês"})
print(response)

response = chain.invoke({"texto": texto, "idioma":"ingles"})
print(response)

response = chain.invoke({"texto": texto, "idioma":"frances"})
print(response)

response = chain.invoke({"texto": texto, "idioma":"português do Brasil"})
print(response)
```
