#### CharachterTextSpliter
Separador simples, por um unico tipo de separador

```python
from langchain_text_splitters import CharacterTextSplitter
```

```python
chunks_size = 50
chunks_overlap = 10
separator = ""


char_splitter = CharacterTextSplitter(
    chunk_size=chunks_size,
    chunk_overlap=chunks_overlap,
    separator=separator
)

char_splitter
```

```python
from string import ascii_lowercase

text_any = ascii_lowercase * 5
text_any, len(text_any)
```

```python
splits = char_splitter.split_text(text_any)
splits, len(splits)
```

```python
text_pascoa = """A Páscoa é uma das celebrações mais importantes do calendário cristão, marcada por um profundo significado espiritual e simbólico. Ela comemora a ressurreição de Jesus Cristo, representando a vitória da vida sobre a morte e trazendo uma mensagem de esperança, renovação e fé para milhões de pessoas ao redor do mundo. Mais do que uma data religiosa, a Páscoa também convida à reflexão sobre valores como amor, solidariedade e perdão."""
len(text_pascoa)
```

```python
chunks_size = 50
chunks_overlap = 20
separator = " "


char_splitter = CharacterTextSplitter(
    chunk_size=chunks_size,
    chunk_overlap=chunks_overlap,
    separator=separator
)
```

```python
char_splitter.split_text(text_pascoa)
```

##### RecursiveCharacterTextSplitter
Separador com mais de um tipo de caracter

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter


chunks_size = 50
chunks_overlap = 20
separator = [" ", ".", ","]

recursive_splitter = RecursiveCharacterTextSplitter(
    chunk_size=chunks_size,
    chunk_overlap=chunks_overlap,
    separators=separator
)
```

```python
recursive_splitter.split_text(text_any)
```

```python
recursive_splitter.split_text(text_pascoa)
```

##### TokenTextSpliter
Separador pela quantidade de tokens

```python
from langchain_text_splitters import TokenTextSplitter


chunks_size = 50
chunks_overlap = 20

token_splitter = TokenTextSplitter(
    chunk_size=chunks_size,
    chunk_overlap=chunks_overlap,
)
```

```python
token_splitter.split_text(text_any)
```

#### SplitTextMarkDown
Uma forma de fazer o split em conteudo markdown, esse tipo de formatação é bem comum para IA

```python
from langchain_text_splitters import MarkdownHeaderTextSplitter


headers = [
    ("#", "Header 1"),
    ("##", "Header 2"),
    ("###", "Header 3"),
]

markdown_splitter = MarkdownHeaderTextSplitter(
    headers_to_split_on=headers
)

```

```python
text_markdown = '''# Titulo do Markdown de exemplo

## Capitulo 1
Texto capitulo 1 e mais e mais texto.
Continuamos no capitulo 1!

## Capitulo 2
Texto capitulo 2 e mais e mais texto.
Continuamos no capitulo 2!

## Capitulo 3
### Secao 3.1
Texto capitulo 3 e mais e mais texto.
Continuamos no capitulo 3!
'''
```

```python
splits = markdown_splitter.split_text(text_markdown)
splits
```

```python
for split in splits:
    print(split.page_content)
    print(split.metadata)
    print("-"*100)
```

#### Split de Documentos
A ideia se repete para documentos

```python
from langchain_community.document_loaders.pdf import PyPDFLoader
from os import path
```

```python
path_file = path.join('/', 'tmp', 'Explorando o Universo das IAs com Hugging Face.pdf')
pdf = PyPDFLoader(path_file)
docs = pdf.load()

len(docs)
```

```python
chunks_size = 50
chunks_overlap = 10
separator = [" ", ".", ","]

recursive_splitter = RecursiveCharacterTextSplitter(
    chunk_size=chunks_size,
    chunk_overlap=chunks_overlap,
    separators=separator
)
```

```python
splits = recursive_splitter.split_documents(docs)
len(splits)
```
