##### Runnables ou Metodos que são invocados para executar em um chain
Métodos dos Runnables de LCEL

A interface padrão de LCEL inclui os seguintes métodos:

- stream: transmitir de volta fragmentos da resposta
- invoke: chamar a cadeia com um input
- batch: chamar a cadeia com uma lista de inputs

```python
from langchain_community.llms import Ollama
from langchain_core.prompts import ChatPromptTemplate
from langchain_core.output_parsers import StrOutputParser
```

```python
chat_model = Ollama(model="minimax-m2.5:cloud")
```

```python
prompt = ChatPromptTemplate.from_template("Crie uma historia sobre a graça do assunto: {assunto}")

chain = prompt | chat_model | StrOutputParser()
```

##### Invokes

```python
response = chain.invoke({"assunto": "Jesus"})
print(response)
```

#### Stream
Para recebermos uma saida conforme ela é gerada pelo modelo utilizamos o __stream__

```python
for chunk in chain.stream({"assunto": "Jesus"}):
    print(chunk)
```

##### Batch
Podemos fazer um lote que pode executar em paralelo

```python
for chunk in chain.batch(["Jesus", "amor", "paz"], config={"configurable": {"max_concurrency": 2}}):
    print(chunk)
```

#### Execução paralela
- Na celula acima vimos que podemos executar com limite de paralelismo, mas ainda não vimos como executar da mesma forma em paralelo.
- Vamos aproveitar tambem e pegar a saida para combinar com outra chain.

```python
from langchain_core.runnables import RunnableParallel
```

```python
prompt = ChatPromptTemplate.from_template("Crie um nome para o seguinte produto: {produto}")

nome_produto = prompt | chat_model | StrOutputParser()
```

```python
prompt = ChatPromptTemplate.from_template("Escreva clientes potenciais para o seguinte produto: {produto}")

clientes_potenciais = prompt | chat_model | StrOutputParser()
```

```python
paralelo = RunnableParallel({"produto": nome_produto, "clientes_potenciais": clientes_potenciais})

paralelo.invoke({"produto": "notebook", "clientes_potenciais": "alunos"})
```

```python
prompt = ChatPromptTemplate.from_template("Crie uma breve historia para o seguinte produto: {produto} e os clientes potenciais: {clientes_potenciais}")

chain = paralelo | prompt | chat_model | StrOutputParser()

response = chain.invoke({"produto": "notebook", "clientes_potenciais": "alunos"})
print(response)
```

#### RunnableLambda
O RunnableLambda permite que você crie um Runnable a partir de uma função Python arbitrária. Isso é útil para encapsular qualquer lógica que você queira aplicar aos dados ou para transformar entradas em saídas de uma maneira específica. Um aspecto interessante do RunnableLambda é que ele pode ser construído para funcionar de forma síncrona ou assíncrona, dependendo de como a função é definida

```python
from langchain_core.runnables import RunnableLambda

chain = RunnableLambda(lambda x: x["produto"]) | nome_produto

chain.invoke({"produto": "notebook", "clientes_potenciais": "alunos"})
```


#### RunnablePassthrough
Esse proximo runnables não faz nada, apenas repassa para o proximo pipe do fluxo

```python
from langchain_core.runnables import RunnablePassthrough

prompt = ChatPromptTemplate.from_template("Esse aqui não é para fazer nada !!! Apenas passar")

paralelo = RunnablePassthrough()

chain = paralelo | prompt | chat_model | StrOutputParser()

chain.invoke({"produto": "notebook", "clientes_potenciais": "alunos"})
```
