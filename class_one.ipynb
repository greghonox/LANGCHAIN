##### ChatModels
###### Agora, vamos explorar os ChatModels, que são uma evolução dos LLMs, projetados especificamente para conversações. Eles entendem e interpretam a conversa, funcionando com três tipos principais de mensagens: __de sistema, de usuário e da IA__.

```python
chat_model = Ollama(model="minimax-m2.5:cloud", temperature=0)

response = chat_model.invoke("Me explique como funciona o LLM?")
print(response)
```

##### HumanMessage: Representa uma mensagem do usuário.
- __AIMessage__: Representa uma resposta do modelo.
- __SystemMessage__: Indica ao modelo como se comportar.
- __FunctionMessage__: Resultado de uma chamada de função específica.
- __ToolMessage__: Resultado de uma chamada de ferramenta específica.

```python
mensagens = [
    SystemMessage(content='Você é um assistente que conta piadas.'),
    HumanMessage(content='Quanto é 1 * 1 + 6?')
]
resposta = chat_model.invoke(mensagens)
print(resposta)
```

```python
mensagens = "Preciso que me conte uma piada sobre o sonic programador"
for trecho in chat_model.stream(mensagens):
    print(trecho, end='')
```

#### Prompt Few-Shot
##### Uma técnica essencial que já discutimos em nosso curso de Engenharia de Prompt é o __Prompt Few-Shot__. Essa técnica envolve fornecer ao modelo exemplos específicos de perguntas e respostas para moldar a maneira como ele responde a consultas futuras. Isso é particularmente útil para alinhar as respostas do modelo com o tipo de saída que desejamos.

##### Vejamos um exemplo prático utilizando o ChatOpenAI da LangChain:

```python
from langchain_core.messages import AIMessage
```

```python
mensagens = [
    HumanMessage(content='Quanto é 1 + 1?'),
    AIMessage(content='2'),
    HumanMessage(content='Quanto é 10 * 5?'),
    AIMessage(content='50'),
    HumanMessage(content='Quanto é 10 + 3?'),
]

chat_model.invoke(mensagens)
```

#### Neste exemplo, alternamos entre HumanMessage e AIMessage para treinar o modelo sobre como esperamos que ele responda. Isso é crucial para garantir que o modelo não só entenda a pergunta, mas também responda no formato desejado.

```python
person = SystemMessage(content='Você é um assistente que conta piadas, mas as vezes muito violento e agressivo.')

response = chat_model.stream([
    person,
    "Me explique como um palhaço se comporta?",
    "Você é um contador de piadas?",
    "Me diga como ser engraçado, como você?!"
])
for trecho in response:
    print(trecho, end='')
```

```python
train_few_shot = [
    person,
    HumanMessage(content='Como se chama ?'),
    AIMessage(content='O sem graça assasino'),
    HumanMessage(content='O que gosta de fazer?'),
    AIMessage(content='Brigar com quem fica me perguntando coisas estúpidas!!!'),
    HumanMessage(content='Quem é o autor da piada?'),    
]

print(chat_model.invoke(train_few_shot))
print(chat_model.invoke([
    person, 
    HumanMessage(content='Como se chama?')]))
```

##### Template Compose

```python
from random import choice

language = choice(["Português", "Japones", "Inglês", "Macedônio"])

template_word_count = ChatPromptTemplate.from_template("""Responda a pergunta em até {tamanho} palavras""", partial_variables={"tamanho": 100})
template_idioms = ChatPromptTemplate.from_template("""Responda no idioma {idioma}""")

template_compose = template_word_count + template_idioms + "Responda a pergunta seguinte seguindo as instruções: {instruction}"
chat = template_compose.format(tamanho=100, idioma=language, instruction="Qual é a capital do Brasil?")

print(chat)
print(chat_model.invoke(chat))
```

```python
from langchain_core.prompts.few_shot import FewShotPromptTemplate
from langchain_core.prompts import PromptTemplate


prompt_template_animal_mamifero = PromptTemplate(
    input_variables=["pergunta", "resposta"],
    template="""
    Pergunta: {pergunta}\nResposta: {resposta}
    """
)

# FewShotPromptTemplate espera lista de dicionários (variáveis do example_prompt)
alguns_exemplos = [
    {"pergunta": "Mamiferos mama quando são filhotes?", "resposta": "Sim"},
    {"pergunta": "Mamiferos normalmente vivem em grupos?", "resposta": "Sim"},
    {"pergunta": "Mamiferos são animais de estimação?", "resposta": "Sim"},
    {"pergunta": "Mamiferos voam?", "resposta": "Não"},
    {"pergunta": "Mamiferos vivem em água?", "resposta": "Não"},
]
alguns_exemplos
```

```python
# FewShotPromptTemplate exige lista de dicts (não strings)
examples_few_shot = [
    {"pergunta": "Mamiferos mama quando são filhotes?", "resposta": "Sim"},
    {"pergunta": "Mamiferos normalmente vivem em grupos?", "resposta": "Sim"},
    {"pergunta": "Mamiferos são animais de estimação?", "resposta": "Sim"},
    {"pergunta": "Mamiferos voam?", "resposta": "Não"},
    {"pergunta": "Mamiferos vivem em água?", "resposta": "Não"},
]

exemplos = FewShotPromptTemplate(
    examples=examples_few_shot,
    example_prompt=prompt_template_animal_mamifero,
    suffix="Pergunta: {input}",
    input_variables=["input"],
)

print(exemplos)
# Para usar o método invoke() do chat_model com FewShotPromptTemplate, 
# você deve primeiro gerar o texto final do prompt usando exemplos.format, 
# passando a variável de input, e então passar esse texto para invoke():

prompt_final = exemplos.format(input="Mamiferos têm pelos?")
print(chat_model.invoke(prompt_final))

prompt_final = exemplos.format(input="Morcegos são mamiferos?")
print(chat_model.invoke(prompt_final))

prompt_final = exemplos.format(input="Mamiferão são animais marinhos?")
print(chat_model.invoke(prompt_final))

prompt_final = exemplos.format(input="Mamiferos são animais de estimação?")
print(chat_model.invoke(prompt_final))


```

```python
from typing import Optional
from pydantic import BaseModel, Field


class Piadas(BaseModel):
    """Uma estrutura para armazenar uma piada"""
    introducao:  str = Field(description="A introdução da piada")
    desenvolvimento: str = Field(description="O desenvolvimento da piada")
    conclusao: str = Field(description="A conclusão da piada")
    avaliacao: Optional[int] = Field(description="A avaliação da piada")
    
    def __str__(self):
        return f"{self.introducao}\n{self.desenvolvimento}\n{self.conclusao}\nAvaliação: {self.avaliacao}"
    
    def __repr__(self):
        return f"Piada(introducao={self.introducao}, desenvolvimento={self.desenvolvimento}, conclusao={self.conclusao}, avaliacao={self.avaliacao})"
```

```python
from langchain_core.output_parsers import PydanticOutputParser
from langchain_core.prompts import ChatPromptTemplate

chat_model = Ollama(model="minimax-m2.5:cloud", temperature=0)

# Ollama (LLM) não implementa with_structured_output; usar parser na chain
parser_piada = PydanticOutputParser(pydantic_object=Piadas)
prompt_piada = ChatPromptTemplate.from_messages([
    ("human", "Conte uma piada conforme o pedido. Responda em JSON "\
    "válido com as chaves: introducao, desenvolvimento, conclusao, "\
    "avaliacao (número 1-10).\n\nPedido: {pedido}\n\n{format_instructions}")
]).partial(format_instructions=parser_piada.get_format_instructions())

chain_piada = prompt_piada | chat_model | parser_piada
resultado = chain_piada.invoke({"pedido": "Conte uma piada sobre um gato"})

print(resultado.introducao)
print(resultado.desenvolvimento)
print(resultado.conclusao)
print(resultado.avaliacao)
print(f"-" * 100)
print(resultado)





```
